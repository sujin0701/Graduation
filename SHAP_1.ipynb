```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/soojin/Desktop/졸작준비/uci-secom.csv')
data.head()
```

```python
data.shape
```

```python
# np.NaN이 결측값임. 이것을 replace을 사용해서 999으로 바꿈
data = data.replace(np.NaN, 999)

# 결측값 정보를 출력
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.drop(columns = ['Time'], axis = 1)
data
```

```python
data[:525]
```

```python
# 예측해야 할 변수인 `Pass/Fail`를 제거하여 머신러닝 입력값인 x에 저장합니다.
# data에는 'Pass/Fail'의 없어집니다. 
x = data.drop(columns = ['Pass/Fail'], axis = 1)

# 예측해야 할 변수 `Pass/Fail`만을 선택하여 numpy 형태로 y에 저장합니다.
y = data['Pass/Fail']

# ravel은 "풀다"로 다차원을 1차원으로 푸는 것을 의미합니다.
# 1차원 벡터 형태로 출력하기 위해 ravel 사용합니다. 
y = y.to_numpy().ravel() 
y
```

```python
# splitting them into train test and split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

# gettiing the shapes
print("shape of x_train: ", x_train.shape)
print("shape of x_test: ", x_test.shape)
print("shape of y_train: ", y_train.shape)
print("shape of y_test: ", y_test.shape)
```
